```python
num_rows = 1000 # the number of rows in our output dataset
outfile_name = PROJECT + '_toy.csv' # the name of the file we're writing to, should end in '.csv'

def genRange(freqs):
    #freqs is array of frequences
    #return array starting at 0, ending at 100, with intervals in between
    #ie: first freq is in range [0] < [1]; second freq is [1] < [2]
    freqs = list(map(lambda x: x*100, freqs))
    intervals = [0]
    for x in freqs:
        intervals.append(x + intervals[-1])
    return intervals


#grab labels from json
labels = list(correlation_values.keys())

#gather frequencies from JSON so I can convert to range
obj_freqs = []
for label in labels:
    obj_freqs.append(correlation_values[label]["freq"])

#print(obj_freqs)

#generate ranges from frequences
obj_ranges = genRange(obj_freqs)
#print(obj_ranges)

count = 0;

##ALGRORITHM:
##FIRST: generate all of the Rows based on the frequency for each label. In other words, generate outcomes first
##SECOND: go through each generated label and generate the rest of the inputs that could generate that output, using the frequencies for each question
##Sub-Algorithm - generating inputs:
##  Turn the frequencies into a range 1-100 (ie: 0.3, 0.7 -> [0, 30, 100])
##  Generate a random int between 1-100
##  Figure out which range it fits in, then assign the label that goes with that range
## ie: if (yes, no) --> (0.3, 0.7) ==> Range is [0, 30, 100] ==> generate random num (like 35) ==> 30 < 35 <= 100 ==> assign no
##Repeat this a bunch, then add to row

##Given a frequency range [0, 25, 55, 60, 100] and labels ["A", "B", "C", "D"] --> return a label based on the frequencies            
##Note: frequency range is always 1 longer than labels.
##  For example: label[0] corresponds to a value falling between range[0] and range[1]
def genValue(ranges, labels):
    val = randint(0, 100)
    #go through ranges object
    for i in range(0, len(ranges) - 1):
        #check if random num is in a certain range
        if(ranges[i] < val and val <= ranges[i+1]):
            #if so: print the label that corresponds to that range
            return labels[i]
    #if something weird happens with randomness and if-statement above never procs, re-run this function recursively to generate new value
    #THIS IS JANKY BUT DON'T KNOW BETTER WAY TO FIX NOW
    return genValue(ranges, labels)


##Writes the file at very end of the process
def writeFile():
    with open(outfile_name, 'w', newline='') as outfile:
        writer = csv.writer(outfile, delimiter=',')
        headers = [field['id'] for field in data_descriptor_json['fields']]
        elt = headers.pop(0)
        headers.append(elt)
        writer.writerow(headers)
        #writer.writerow(labels)

        for row in rows:
            writer.writerow(row)

#This part generates the output labels using frequencies.
#There's a weird thing with the loop & probability that I couldn't figure out, so using a while look to make sure I generate the right count
tempLabels = []
while(count < num_rows):
    val = randint(0, 100)
    #go through ranges object
    for i in range(0, len(obj_ranges) - 1):
        #check if random num is in a certain range
        if(obj_ranges[i] < val and val <= obj_ranges[i+1]):
            #if so: print the label that corresponds to that range
            tempLabels.append(labels[i])
            count += 1
            
#tempLabels now has num_rows listings of labels distributed by frequency
#print("length of labels: " + str(len(tempLabels)))

##This part goes through each output label and generates the rest of the inputs according to the probabilities associated with that output
rows = []
for label in tempLabels:
    #print(label)
    tempRow = []
    obj = correlation_values[label]
    for field in obj["correlated_fields"]:
        #print(field["name"]);
        category_values = field["correlated_category_values"]
        #print(field)
        objLabels = []
        objFreqs = []
        for key in category_values:
            objLabels.append(key);
            objFreqs.append(category_values[key])
            #print("\t " + key +": " + str(category_values[key]))
            #print("\t" + str(values))
        objRanges = genRange(objFreqs)
        tempLabel = genValue(objRanges, objLabels)
        tempRow.append(tempLabel)
    tempRow.append(label)
    #print(tempRow)
    rows.append(tempRow)

print("rows: " + str(len(rows)))
for row in rows:
    print(row)



writeFile()
            
            
#print(tempLabels)
#print(len(tempLabels))
#for label in labels:
#    print(tempLabels.count(label) / float(num_rows))
    


#test = [0, 0.1, 0.8, 0.1]
#newTest = genRange(test)
#print(newTest)
```
